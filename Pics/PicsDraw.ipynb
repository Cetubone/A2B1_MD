## ViolinBox

```python
import pandas as pd
RMSD_dat = pd.DataFrame({
    "PAC5" : y_p4c5,
    "RNA" : y_p4rna,
    "ssDNA" : y_2p42dna
}
)
RMSD_dat.reset_index(inplace = True)
RMSD_dat.columns = ["FrameNdx","PAC5", "RNA", "ssDNA"]
RMSD_dat
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm # type: ignore
from statsmodels.formula.api import ols # type: ignore
from statannotations.Annotator import Annotator # type: ignore

df = RMSD_dat

# 
columns_to_plot = ['PAC5', 'RNA', 'ssDNA']

# 
df_long = df.melt(id_vars=['FrameNdx'], value_vars=columns_to_plot, var_name='Variable', value_name='Value')

PAC5_col = "#2F81B9"
RNA_col = "#FF9333"
ssDNA_col = "#3AA63A"

# 
plt.figure(figsize=(6, 6), dpi=300)
ax = sns.violinplot(
    x='Variable',
    y='Value',
    data=df_long,
    hue='Variable',  #  hue 
    palette={'PAC5': PAC5_col, 'RNA': RNA_col, 'ssDNA': ssDNA_col},
    inner=None,  # 
    linewidth=0,
    density_norm="width",  #  density_norm 
    cut=0,  # 
    alpha=0.7,
    legend=False  # 
)

# 
sns.boxplot(
    x='Variable',
    y='Value',
    data=df_long,
    ax=ax,
    showfliers=False,  # 
    linewidth=1,  # 
    width=0.3,  # 
    hue='Variable',  #  hue 
    palette={'PAC5': PAC5_col, 'RNA': RNA_col, 'ssDNA': ssDNA_col},
    zorder=10,  # 
    legend=False  # 
)

# 
plt.title('RMSD')
plt.xlabel(' ')
plt.ylabel('RMSD (nm)')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.grid(True, axis="y")

# ANOVA
model = ols('Value ~ C(Variable)', data=df_long).fit()
anova_results = sm.stats.anova_lm(model, typ=2)
print(anova_results)

# 
pairs = [
    ('PAC5', 'RNA'),
    ('PAC5', 'ssDNA'),
    ('RNA', 'ssDNA')
]

annotator = Annotator(ax, pairs, data=df_long, x='Variable', y='Value')
annotator.configure(test='t-test_ind', text_format='star', loc='inside')
annotator.apply_and_annotate()

# 
plt.savefig("RMSD_ViolinBox.svg")
plt.savefig("RMSD_ViolinBox.png", dpi=300)

# 
plt.show()
plt.close('all')
```

```python
import pandas as pd
RMSF_dat = pd.DataFrame({
    "PAC5" : y_p4c5,
    "RNA" : y_p4rna,
    "ssDNA" : y_2p42dna
}
)
RMSF_dat.reset_index(inplace = True)
RMSF_dat.columns = ["FrameNdx","PAC5", "RNA", "ssDNA"]
RMSF_dat
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm # type: ignore
from statsmodels.formula.api import ols # type: ignore
from statannotations.Annotator import Annotator # type: ignore

df = RMSF_dat

# 
columns_to_plot = ['PAC5', 'RNA', 'ssDNA']

# 
df_long = df.melt(id_vars=['FrameNdx'], value_vars=columns_to_plot, var_name='Variable', value_name='Value')

PAC5_col = "#2F81B9"
RNA_col = "#FF9333"
ssDNA_col = "#3AA63A"

# 
plt.figure(figsize=(6, 6), dpi=300)
ax = sns.violinplot(
    x='Variable',
    y='Value',
    data=df_long,
    hue='Variable',  #  hue 
    palette={'PAC5': PAC5_col, 'RNA': RNA_col, 'ssDNA': ssDNA_col},
    inner=None,  # 
    linewidth=0,
    density_norm="width",  #  density_norm 
    cut=0,  # 
    alpha=0.7,
    legend=False  # 
)

# 
sns.boxplot(
    x='Variable',
    y='Value',
    data=df_long,
    ax=ax,
    showfliers=False,  # 
    linewidth=1,  # 
    width=0.3,  # 
    hue='Variable',  #  hue 
    palette={'PAC5': PAC5_col, 'RNA': RNA_col, 'ssDNA': ssDNA_col},
    zorder=10,  # 
    legend=False  # 
)

# 
plt.title('Radius of gyration (total and around axes)')
plt.xlabel(' ')
plt.ylabel('Rg (nm)')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
plt.grid(True, axis="y")

# ANOVA
model = ols('Value ~ C(Variable)', data=df_long).fit()
anova_results = sm.stats.anova_lm(model, typ=2)
print(anova_results)

# 
pairs = [
    ('PAC5', 'RNA'),
    ('PAC5', 'ssDNA'),
    ('RNA', 'ssDNA')
]

annotator = Annotator(ax, pairs, data=df_long, x='Variable', y='Value')
annotator.configure(test='t-test_ind', text_format='star', loc='inside')
annotator.apply_and_annotate()

# 
plt.savefig("Gyrate_ViolinBox.svg")
plt.savefig("Gyrate_ViolinBox.png", dpi=300)

# 
plt.show()
plt.close('all')
```
